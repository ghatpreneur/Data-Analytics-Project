```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

cmaps = [
    ('Perceptually Uniform', ['viridis', 'plasma', 'inferno', 'magma', 'cividis']),
    ('Sequential', ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                    'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                    'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn']),
    ('Diverging', ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu',
                   'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic']),
    ('Qualitative', ['Pastel1', 'Pastel2', 'Paired', 'Accent',
                     'Dark2', 'Set1', 'Set2', 'Set3',
                     'tab10', 'tab20', 'tab20b', 'tab20c'])
]

gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

def plot_color_gradients(cmap_category, cmap_list):
    fig, axes = plt.subplots(nrows=len(cmap_list), figsize=(7, 0.3 * len(cmap_list) + 1))
    fig.subplots_adjust(top=0.95, bottom=0.01, left=0.2, right=0.99)
    axes[0].set_title(cmap_category + ' Colormaps', fontsize=14)

    for ax, name in zip(axes, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
        ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10, transform=ax.transAxes)
        ax.set_axis_off()

for cmap_category, cmap_list in cmaps:
    plot_color_gradients(cmap_category, cmap_list)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches


palettes = {
    "1. Modern Flat UI": {
        'Midnight Blue (#2C3E50)': '#2C3E50',
        'Alizarin Red (#E74C3C)': '#E74C3C',
        'Peter River (#3498DB)': '#3498DB', 
        'Emerald Green (#2ECC71)': '#2ECC71',
        'Sun Flower (#F1C40F)': '#F1C40F',
        'Amethyst (#9B59B6)': '#9B59B6',
        'Carrot (#E67E22)': '#E67E22',
        'Concrete (#95A5A6)': '#95A5A6'
    },
    "2. Risk Levels (Google Material)": {
        'Critical (#D32F2F)': '#D32F2F',
        'High (#F44336)': '#F44336',
        'Medium (#FF9800)': '#FF9800', 
        'Low (#FFEB3B)': '#FFEB3B',
        'Safe (#4CAF50)': '#4CAF50',
        'Neutral (#9E9E9E)': '#9E9E9E'
    },
    "3. Colorblind Safe (Okabe-Ito)": {
        'Orange (#E69F00)': '#E69F00',
        'Sky Blue (#56B4E9)': '#56B4E9',
        'Bluish Green (#009E73)': '#009E73',
        'Yellow (#F0E442)': '#F0E442',
        'Blue (#0072B2)': '#0072B2',
        'Vermilion (#D55E00)': '#D55E00',
        'Purple (#CC79A7)': '#CC79A7'
    }
}

def plot_hex_palettes(palette_dict):
    n_groups = len(palette_dict)
    fig, axes = plt.subplots(n_groups, 1, figsize=(8, 3 * n_groups))
    
   
    if n_groups == 1: axes = [axes]
    
    for ax, (title, colors) in zip(axes, palette_dict.items()):
        ax.set_title(title, fontsize=14, fontweight='bold', loc='left', pad=10)
        ax.set_xlim(0, 10)
        ax.set_ylim(0, len(colors))
        ax.axis('off')
        
        for i, (name, hex_code) in enumerate(colors.items()):
          
            rect = patches.Rectangle((0, i), 2, 0.8, linewidth=1, edgecolor='none', facecolor=hex_code)
            ax.add_patch(rect)
            
            
            text_color = 'black'
            ax.text(2.2, i + 0.4, f"{name}", va='center', fontsize=11, fontweight='bold', color='#333333')
            
    plt.tight_layout()
    plt.show()

plot_hex_palettes(palettes)
```
